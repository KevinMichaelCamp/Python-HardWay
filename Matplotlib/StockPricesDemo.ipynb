```python
import matplotlib.pyplot as plt
%matplotlib inline
from matplotlib.ticker import MaxNLocator
```

```python
apl_price = [93.95, 112.15, 104.05, 144.85, 169.49]
ms_price = [39.01, 50.29, 57.05, 69.98, 94.39]
year = [2014, 2015, 2016, 2017, 2018]
```

```python
plt.plot(year, apl_price, 'b')
plt.plot(year, ms_price, 'g')
plt.xlabel('Year')
plt.ylabel('Stock Price')
plt.axis([2014, 2018, 0, 200])
plt.show()
```

```python
fig_1 = plt.figure(1, figsize=(20, 5))
apple_chart = fig_1.add_subplot(121)
plt.xlabel('Year')
plt.ylabel('Stock Price')
microsoft_chart = fig_1.add_subplot(122)
plt.xlabel('Year')
plt.ylabel('Stock Price')

apple_chart.plot(year, apl_price, 'g')
apple_chart.set_title('Apple')
apple_chart.xaxis.set_major_locator(MaxNLocator(integer=True))
apple_chart.axis([2014,2018,0,200])

microsoft_chart.plot(year, ms_price, 'b')
microsoft_chart.set_title('Microsoft')
microsoft_chart.xaxis.set_major_locator(MaxNLocator(integer=True))
microsoft_chart.axis([2014,2018,0,200])

plt.show()
```

```python
fig_2, axes = plt.subplots(1,2,figsize=(20,5))
plt.ylabel('Stock Price')
plt.xlabel('Year')
axes[0].plot(year, apl_price, 'g')
axes[1].plot(year, ms_price, 'b')
axes[0].xaxis.set_major_locator(MaxNLocator(integer=True))
axes[1].xaxis.set_major_locator(MaxNLocator(integer=True))
axes[0].set_title('Apple')
axes[1].set_title('Microsoft')
axes[0].axis([2014,2018,0,200])
axes[1].axis([2014,2018,0,200])

plt.show()
```
