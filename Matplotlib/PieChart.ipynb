```python
import matplotlib.pyplot as plt
```

```python
labels = 'Bike', 'Gym', 'Work', 'Play'
sizes = [5,10,45,40]
explode = (0, 0, 0.1, 0)

fig1, ax1 = plt.subplots(figsize=(20,10,))
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', shadow=True, startangle=90)
ax1.axis('equal')
plt.show()
```
