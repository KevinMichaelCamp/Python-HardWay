```python
import numpy as np
```

```python
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 20)
```

```python
y1 = x**2
```

```python
y2 = x**1.5
```

```python
plt.plot(x, y1, "bo-", linewidth=.5, markersize=2, label= "Squared")
plt.plot(x, y2, "gs-", linewidth=.5, markersize=2, label= "1.5")
plt.xlabel("$X-Axis$")
plt.ylabel("$Y-Axis$")
plt.legend(loc="upper left")
plt.savefig("myplot.png")

```
