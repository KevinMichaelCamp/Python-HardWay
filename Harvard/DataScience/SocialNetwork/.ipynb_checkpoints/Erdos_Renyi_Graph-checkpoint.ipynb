```python
import matplotlib.pyplot as plt
from scipy.stats import bernoulli
import networkx as nx
```

```python
bernoulli.rvs(p=0.2)
```

```python
N = 20
p = 0.2

def er_graph(N, p):
    """Generate an ER grph"""
    #create empty graph
    G = nx.Graph()
    #add all N nodes in the graph
    G.add_nodes_from(range(N))
    # loop all all pairs of nodes
    for node1 in G.nodes():
        for node2 in G.nodes():
            #add an edge with probability p
            if node1 < node2 and bernoulli.rvs(p=p):
                G.add_edge(node1, node2)
                
    return G
    
```

```python
nx.draw(er_graph(50, 0.08), with_labels=True, node_color='lightblue', edge_color='grey')
```

```python
def plot_degree_distribution(G):
    plt.hist(list(dict(G.degree()).values()), histtype="step")
    plt.xlabel("Degree $k$")
    plt.ylabel("$P(k)$")
    plt.title("Degree Distribution")
```

```python
G1 = er_graph(500, 0.08)
G2 = er_graph(500, 0.08)
G3 = er_graph(500, 0.08)
```

```python
plot_degree_distribution(G1)
plot_degree_distribution(G2)
plot_degree_distribution(G3)
```
