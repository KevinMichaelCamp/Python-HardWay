```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#random walks
```

```python
X_0 = np.array([[0], [0]])
deltaX = np.random.normal(0,1,(2,10000))
X = np.concatenate((X_0, np.cumsum(deltaX, axis = 1)), axis = 1)
plt.plot(X[0], X[1], "bo-", linewidth=.5, markersize=2);
```
